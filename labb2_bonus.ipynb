```python
import pandas as pd
df2 = pd.read_excel("betyg_o_prov_riksnivå.xlsx",
                   sheet_name="Tabell 1B",
                   skiprows=7)

df2
```

```python
df2 = df2.rename(columns={'Unnamed: 0' : 'År/Huvud/Bak'}) # Denna funkar, inte genom index
# df2 = df2.rename(columns={'År' : 'År/Huvud/Bak'}) # Denna funkar, inte genom index

df2
```

```python
import matplotlib.pyplot as plt

# Hitta rätt rader för kommunala skolor och waldorfskolor
kommunal_rad = df2[df2['År/Huvud/Bak'] == 'Kommun'].iloc[0]
waldorf_rad = df2[df2['År/Huvud/Bak'] == 'Waldorf'].iloc[0]
konfessionell_rad = df2[df2['År/Huvud/Bak'] == 'Konfessionell'].iloc[0]
allmäna_friskor_rad = df2[df2['År/Huvud/Bak'] == 'Allmän inriktning'].iloc[0]

# Hämta värdena för ofullständiga betyg (Totalt.2 kolumnen)
kommunal_ofullstandiga = kommunal_rad['Totalt.2']
waldorf_ofullstandiga = waldorf_rad['Totalt.2']
konfessionell_ofullständiga = konfessionell_rad['Totalt.2']
allmäna_friskolor_ofullständiga = allmäna_friskor_rad['Totalt.2']

# Skapa barplot
skoltyper = ['Kommunala', 'Konfessionella', 'Waldorf', 'Allmän inriktning']
ofullstandiga = [kommunal_ofullstandiga, konfessionell_ofullständiga, waldorf_ofullstandiga, allmäna_friskolor_ofullständiga]

plt.figure(figsize=(10, 6))
plt.bar(skoltyper, ofullstandiga)
plt.title('Jämförelse av ofullständiga betyg i 9:e klass')
plt.xlabel('Skoltyp')
plt.ylabel('Andel ofullständiga betyg (%)')
plt.ylim(0, max(ofullstandiga) * 1.1)  # Sätter y-axelns övre gräns till 110% av det högsta värdet

for i, v in enumerate(ofullstandiga):
    plt.text(i, v, f'{v:.1f}%', ha='center', va='bottom')
    
# Spara figuren
plt.savefig('Visualiseringar/Jämförelse_ofullständiga_betyg.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Filtrera vad som ska finnas med på x-axeln
skoltyper = ['Kommun', 'Konfessionell', 'Waldorf', 'Allmän inriktning']
# Sortera och bestäm y-axeln
data = df2[df2['År/Huvud/Bak'].isin(skoltyper)].set_index('År/Huvud/Bak')['Totalt.2']

# Skapa barplot
plt.figure(figsize=(10, 6))
plt.bar(data.index, data.values)
plt.title('Jämförelse av ofullständiga betyg i 9:e klass')
plt.xlabel('Skoltyp')
plt.ylabel('Andel ofullständiga betyg (%)')

#loopa index och values i data
for i, v in enumerate(data):
    plt.text(i, v, f'{v:.1f}%', ha='center', va='bottom')

# Spara figuren
plt.savefig('Visualiseringar/totalt(poäng)_eng_math_sve_sva.png', dpi=300, bbox_inches='tight')

plt.show()
```

```python
#Filtrera
bakgrunder = ('Elever med svensk bakgrund', 'Elever med utländsk bakgrund', 'Födda i Sverige', 'Nyinvandrade elever')

#Sortera
data = df2[df2['År/Huvud/Bak'].isin(bakgrunder)].set_index('År/Huvud/Bak')['Totalt.2']

#Skapa
plt.figure(figsize=(12, 6))
plt.bar(data.index, data.values)
plt.title('Jämförelse av ofullständiga betyg - bakgrund')
plt.xlabel='Bakgrund'
plt.ylabel='Andel ofullständiga betyg (%)'
plt.ylim(0, data.max() * 1.35)

#Loopa index och värden
for i, v in enumerate(data):
    plt.text(i, v, f'{v:.1f}%', ha='center', va='bottom')

# Spara figuren
plt.savefig('Visualiseringar/Ofullständiga_betyg_bakgrund.png', dpi=300, bbox_inches='tight')

plt.show()
```

# Undersök korrelationen mellan antalet elever med icke godkända betyg och föräldrars utbildningsnivå

1. filtrera efter föräldrars utbildningsgrad
2. sortera efter de som inte har fullständiga betyg
3. skapa plot
4. loopa index och values
5. visa


```python
import matplotlib.pyplot as plt

föräldrar_utbildning = ('Förgymnasial utbildning', 'Gymnasial utbildning', 'Eftergymnasial utbildning')

data3 = df2[df2['År/Huvud/Bak'].isin(föräldrar_utbildning)].set_index('År/Huvud/Bak')['Totalt.2']

plt.figure(figsize=(10, 6))
plt.bar(data3.index, data3.values)
plt.title('Elever med ofullständiga betyg (%) - korrelation föräldrars utbildningsgrad')
plt.xlabel('Föräldrars utbildningsgrad')
plt.ylabel('Andel elever med ofullständiga betyg (%)')
plt.ylim(0, data3.max() * 1.4)

for i, v in enumerate(data3):
    plt.text(i, v, f'{v:.1f}%', ha='center', va='bottom')

plt.savefig('Föräldrars utbildningsgrad för elever med ofullständiga betyg.png', dpi=300, bbox_inches='tight')

plt.show()
```
